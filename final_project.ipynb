```python
#show sum missing data into decimal of other
x = data.isna()
y = x.sum()
z = y / len(data)
z
```

```python
#drop column with > 50% missing value
ms = int(0.5*data.shape[0] + 1)
data = data.dropna(axis = 1, thresh = ms)
```

```python
data.info()
```

```python
#fill missing value to column 'country' and 'children' with mode of data.
data['country'] = data['country'].fillna(data['country'].mode()[0])
data['children'] = data['children'].fillna(data['children'].mode()[0])
data
```

```python
data.isnull().sum()
```

```python
#fill missing value to column 'agent' with mean.
data['agent'] = data['agent'].fillna(data['agent'].mean)
data
```

```python
data.isnull().sum()
```

```python
#Buat kategori berdasarkan lead time: Short, Medium, dan Long
import numpy as np
```

```python
conditions = [(data["lead_time"] > 160),
              (data["lead_time"] <= 160)&(data["lead_time"] >= 69),
             (data["lead_time"] < 69)]
choices = ["Long","Medium","Short"]
data["kategori"] = np.select(conditions,choices)
data
```

```python
#Filter semua data yang memiliki reservation_status_date di > 2017
#mengganti tipe data menjadi datetime
data["reservation_status_date"] = pd.to_datetime(data["reservation_status_date"],format='%Y/%m/%d', errors = 'coerce')
data.info()
```

```python
import datetime as dt

fy = data[data['reservation_status_date'].dt.year >= 2017]
fy
```
