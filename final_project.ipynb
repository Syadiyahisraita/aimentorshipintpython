```python
import pandas as pd #import pandas
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("hotel.csv") #open with pd.read_csv function
data
```

```python
#data.info() #show information of data like column, dtype, missing value, etc.
#properties data 'hotel.csv'
print(data.info())
print()

#jumlah data 
count_null = data.isna().sum().sum()
count_notnull = data.notnull().sum().sum()
print('Jumlah Data Missing Values : ', count_null)
print()
print('List Kolom dengan Data Missing Values :')
print(data.isna().sum())
print()
print('Jumlah Data with Values : ', count_notnull)
print('Jumlah Seluruh Data : ', (count_notnull + count_null))


#grafik dtypes pada data
slices = [4, 16, 12]
labels = ['float', 'int', 'object']

plt.pie(slices, labels=labels, autopct = '%1.1f%%')

plt.title("Data Types")
plt.tight_layout()
plt.show()

#grafik missing values pada data
slices = [488, 16340, 4, 112593]
labels = ['country', 'agent','children', 'company']

plt.pie(slices, labels=labels, autopct = '%1.1f%%')

plt.title("Missing Values")
plt.tight_layout()
plt.show()
```

```python
#data.describe() #show describe count data, mean, median, dan standar deviasi, quantile, min data, max data
#count, mean, std, dan median dari data
data_t = data.describe().T
data_t2 = data_t.loc[:,['count','mean','std','50%']]
data_t2.columns = ['Jumlah Data','Mean','Std','Median']
data_t2
```

```python
#min dan max dari data
data_t = data.describe().T
data_t[['min','max']]
```

```python
#drop kolom dengan 50% lebih missing value

#hitung jumlah threshold missing values (50%)
persen = int(0.5*data.shape[0] + 1)

#cetak grafik kolom vs jumlah missing values
jumlah = [488, 16340, 4, 112593]
kolom = ['country', 'agent','children', 'company']

miss_val = pd.DataFrame({"Columns":kolom,
                  "Jumlah":jumlah})

miss_val = miss_val.sort_values('Jumlah')

plt.bar('Columns', 'Jumlah', data = miss_val)

plt.title("Columns with Missing Values")
plt.axhline(y = persen, linestyle = '--', c = 'r', label='threshold 50% + 1')
plt.xlabel("Columns")
plt.ylabel("Number of Missing Values")
plt.legend()

plt.tight_layout()
plt.show()

#menghapus kolom yg melebihi threshold, kemudian dicek dgn df.info()
data.dropna(axis = 1, thresh = persen, inplace = True)
data.info()
```

```python
#mengisi missing value dengan modus

#mengecek modus dari kolom tertentu
print('modus dari country : ', data['country'].mode()[0])
print('modus dari children : ',data['children'].mode()[0])
print()

#mengisi kolom tersebut dgn modus, kemudian dicek
data['country'] = data['country'].fillna(data['country'].mode()[0])
data['children'] = data['children'].fillna(data['children'].mode()[0])
print(data.isna().sum())
print()

#cetak grafik children
slices = data['children'].value_counts()
labels = data['children'].value_counts().index.tolist()
myexplode = [0, 0, 0, 0.5, 1.5]

plt.pie(slices, labels=labels, explode = myexplode, autopct = '%1.1f%%')

plt.title("children graph")
plt.tight_layout()
plt.show()

#cetak grafik country
print('Plot kolom country menggunakan 5 data terbanyak')
jumlah = data['country'].value_counts().head()
country = data['country'].value_counts().head().index.tolist()

miss_val = pd.DataFrame({"Columns":country,
                  "Jumlah":jumlah})

miss_val = miss_val.sort_values('Jumlah', ascending=False)

plt.bar('Columns', 'Jumlah', data = miss_val)

plt.title("country graph")
plt.xlabel('country')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
#mengisi missing value dengan mean

#mencari mean dari kolom agent
print('mean dari agent : ', data['agent'].mean())
print()

#isi kolom agent dengan mean untuk missing value, kemudian dicek dengan df.isna
data['agent'] = data['agent'].fillna(data['agent'].mean())
print(data.isna().sum())

#cetak grafik agent
print('Plot kolom agent menggunakan 5 data terbanyak')
slices = data['agent'].value_counts().head()
labels = data['agent'].value_counts().head().index.tolist()
myexplode = [0, 0, 0, 0, 0]

plt.pie(slices, labels=labels, explode = myexplode, autopct = '%1.1f%%')

plt.title("agent graph")
plt.tight_layout()
plt.show()

#cetak grafik agent bar
print('Plot kolom agent menggunakan 5 data terbanyak')
jumlah = data['agent'].value_counts().head()
agent = data['agent'].value_counts().head().index.tolist()

miss_val = pd.DataFrame({"Columns":agent,
                  "Jumlah":jumlah})

miss_val = miss_val.sort_values('Jumlah', ascending=False)

plt.bar('Columns', 'Jumlah', data = miss_val)

plt.title("agent graph")
plt.xlabel('agent')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
#Kategorikan lead time, asumsi pembuat coding
conditions = [
    (data['lead_time'] < 69),
    (data['lead_time'] >= 69) & (data['lead_time'] <= 169),
    (data['lead_time'] > 169)
    ]

#mendefinisikan output
values = ['Short', 'Medium', 'Long']

#add kolom baru dengan isi data kondisi
data['Duration'] = np.select(conditions, values)

#cetak grafik duration
slices = data['Duration'].value_counts()
labels = data['Duration'].value_counts().index.tolist()

plt.pie(slices, labels=labels, autopct = '%1.1f%%')

plt.title("Duration")
plt.tight_layout()
plt.show()

data['Duration']
```

```python
#mengganti tipe data menjadi datetime
data["reservation_status_date"] = pd.to_datetime(data["reservation_status_date"],format='%Y/%m/%d', errors = 'coerce')
data["reservation_status_date"].dtype
```
